```python

f = open("17_21416.txt")
# s = list(map(int, f.read().split()))
s = [int(i) for i in f.read().split()]

summ = 0
for i in s:
    if i<0:
        summ+=i

count = 0
mx = -999999999
for i in range(0,len(s)-2):
    arr = [s[i],s[i+1],s[i+2]]
    if max(arr)*min(arr)>summ:
        count+=1
        mx = max(mx, sum(arr))
print(count,mx)




```

```python
import sys

f = open("input.txt")
s = f.readline().strip()

size = 2005
offset = 1002

visited = [[0] * size for _ in range(size)]

x = offset
y = offset
visited[x][y] = 1
ans = 0

for i in range(len(s)):
    if s[i] == 'U':
        y += 1
    elif s[i] == 'D':
        y -= 1
    elif s[i] == 'L':
        x -= 1
    elif s[i] == 'R':
        x += 1

    visited[x][y] += 1
    if visited[x][y] == 2:
        ans += 1

print(ans)
```
